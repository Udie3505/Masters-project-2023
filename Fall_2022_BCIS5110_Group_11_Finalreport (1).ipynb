#### step 6 : To compensate the null values we replace the null values with median in 'HDI for year'

```python
##for replacing the misiing values plotting both mean and median as we can use one of them for replacing the missing values
df['HDI for year'].plot(kind='kde').set(title='Fig 7')
plt.vlines(df['HDI for year'].mean(), ymin= 0, ymax = 40, color = 'r', label = 'mean')
plt.vlines(df['HDI for year'].median(), ymin = 0, ymax = 40, color = 'g', label = 'median')

plt.legend()
```

From the above figure 7 we can see both mean and median lie approximately on the same line we are replacing the missing values by median because as we can see median value is already present in the data set and is less susceptible to outlier errors as compared to mean imputation.

### The below steps have been performed for imputing the missing values

```python
## replacing the null values with median
for col in ['HDI for year']:
    df[col].fillna(df[col].median(), inplace = True)
    
```
